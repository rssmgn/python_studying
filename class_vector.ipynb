```python
class Vector:
    def __init__(self, x, y, z):
        self.x = float(x)
        self.y = float(y)
        self.z = float(z)
        self.hist = []
        
    # сумма
    def __add__ (self, b):
        ax = self.x + b.x
        ay = self.y + b.y
        az = self.z + b.z
        
        self.hist += [("add", (self.x , self.y, self.z), (b.x , b.y, b.z), (ax, ay, az))]
        
        return Vector(ax, ay, az)
    
    # разность
    def __sub__ (self, b):
        sx = self.x + b.x
        sy = self.y + b.y
        sz = self.z + b.z
        
        self.hist += [("sub", (self.x , self.y, self.z), (b.x , b.y, b.z), (sx, sy, sz))]
        
        return Vector(sx, sy, sz)
        
    # скалярное произведение  
    def __mul__(self, b):
        mul = (self.x * b.x) + (self.y * b.y) + (self.z * b.z)
        
        self.hist += [("mul", (self.x , self.y, self.z), (b.x , b.y, b.z), mul)]        
        b.hist += [("mul", (self.x , self.y, self.z), (b.x , b.y, b.z), mul)]
        
        return mul
    
    # векторное произведение 
    def __xor__(self, b):
        xx = (self.y * b.z) - (self.z * b.y)
        xy = (self.z * b.x) - (self.x * b.z)
        xz = (self.x * b.y) - (self.y * b.x)
        
        self.hist += [("xor", (self.x , self.y, self.z), (b.x , b.y, b.z), (xx, xy, xz))]
        b.hist += [("xor", (self.x , self.y, self.z), (b.x , b.y, b.z), (xx, xy, xz))]
        
        return Vector(xx, xy, xz)
    
    # длина вектора
    def lenVector(self):
        res = (self.x ** 2 + self.y ** 2 + self.z ** 2) ** 0.5
        
        self.hist += [("lenVector", (self.x , self.y, self.z), res)]
        
        return res
    
    # проверка на коллинеарность
    def isCollin(self, b):
        v = (self ^ b)
        
        if abs(v.x) + (v.y) + (v.z) == 0:
            res = True
        else:
            res = False
        
        self.hist += [("isCollin", (self.x , self.y, self.z), (b.x , b.y, b.z), res)]
        b.hist += [("isCollin", (self.x , self.y, self.z), (b.x , b.y, b.z), res)]

        return res
    
    # проверка на компланарность
    def isCompl(self, b1, b2):
        v = (self ^ b1) * b2
        
        if v == 0:
            res = True
        else:
            res = False
        
        self.hist += [("isCompl", (self.x , self.y, self.z), (b1.x , b1.y, b1.z), (b2.x , b2.y, b2.z), res)]
        b1.hist += [("isCompl", (self.x , self.y, self.z), (b1.x , b1.y, b1.z), (b2.x , b2.y, b2.z), res)]
        b2.hist += [("isCompl", (self.x , self.y, self.z), (b1.x , b1.y, b1.z), (b2.x , b2.y, b2.z), res)]
        
        return res
    
    # Перегружаем базовую функцию str()
    def __str__(self):
        return str((self.x, self.y, self.z))
    
    def history(self):
        return tuple(self.hist)
    
    def info(self):
        print("Координаты вектора: ", (self.x, self.y, self.z))
        print("Длина вектора: ", self.lenVector())
        print("История вектора: ",self.history())
```

```python
v1 = Vector(1,2,3)
v2 = Vector(4,5,6)
v3 = Vector(7,8,9)

print("Вектор 1: ", v1)
print("Вектор 2: ", v2)
print("Вектор 3: ", v3)
print("Сумма векторов ", v1, "и", v2, "=", v1 + v2)
print("Разность векторов ", v2, "и", v3, "=", v2 - v3)
print("Скалярное произведение векторов ", v1, "и", v2, "=", v1 * v2)
print("Векторное произведение векторов ", v2, "на", v3, "=", v2 ^ v3)
print("Длина вектора: ", v1, "=", v2.lenVector())

if v2.isCollin(v3):
    print("Векторы", v2, "и", v3, "коллинеарны")
else:
    print("Векторы", v2, "и", v3, "не коллинеарны")

if v1.isCompl(v2,v3):
    print("Векторы", v1, ",", v2, ",", v3,"компланарны")
else:
    print("Векторы", v1, ",", v2, ",", v3,"не компланарны")
    
print("История вектора ", v1, "-", v1.history())
print("История вектора ", v2, "-", v2.history())
print("История вектора ", v3, "-", v2.history())
print("Информация о векторе ", v1, ":")
v1.info()
```
